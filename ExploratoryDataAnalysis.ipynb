```python
import pandas as pd
print(dir(pd))
```

```python
path = 'https://raw.githubusercontent.com/a-forty-two/hsbc-gcp-private/main/data.csv'
data = pd.read_csv(path)
```

```python
print(dir(data))
```

```python
data.columns
```

```python
data.tail(6)
```

```python
# SLICING AND DICING (select)
# by position or by name
# by position or index -> iloc
# by name-> loc
# for both loc and iloc, [ row selection, column selection]
data = data.iloc[ : , :-1]
data.tail(6)
```

```python
# alternate way of doing this using loc 
path = 'https://raw.githubusercontent.com/a-forty-two/hsbc-gcp-private/main/data.csv'
data = pd.read_csv(path)

col_names = data.columns
col_names = col_names[:-1]
data =data.loc [ :, col_names]
data.tail(6)
```
